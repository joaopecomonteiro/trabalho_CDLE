```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file_path = "1mes.csv"

df = pd.read_csv(file_path)
df.set_index('Task', inplace=True)



# Filter tasks that do not start with 'filtered'
other_tasks = df[~df.index.str.startswith('filtered')]

# Plotting other tasks
other_tasks.plot(kind='bar', figsize=(12, 8))
plt.title('Standart Tasks')
plt.xlabel('Library')
plt.ylabel('Time')
plt.xticks(rotation=45)
plt.legend(title='Task')
plt.show()

# Filter tasks that start with 'filtered'
filtered_tasks = df[df.index.str.startswith('filtered')]

# Plotting filtered tasks
filtered_tasks.plot(kind='bar', figsize=(12, 8))
plt.title('Filtered Tasks')
plt.xlabel('Library')
plt.ylabel('Time')
plt.xticks(rotation=45)
plt.legend(title='Task')
plt.show()

```
